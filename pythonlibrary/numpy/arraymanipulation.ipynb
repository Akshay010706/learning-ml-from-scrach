# Array Manipulation

```python
import numpy as np
```

```python
array = np.random.randint(10,50,(3,3))
print(array)
print(array.shape)
```

1. Transpose

```python
trans = np.transpose(array)
print(trans)
print(trans.shape)
```

or

```python
tran2 = array.T
print(tran2)
print(tran2.shape)
```

2.Reshaping an array

```python
a = np.random.randint(10,30,(3,4))
print(a)
print(a.shape)
```

```python
b = a.reshape(4,3)
print(b)
print(b.shape)
```
