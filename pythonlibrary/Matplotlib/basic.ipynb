# bar plot

```python
fig = plt.figure() # note: figure(), not Figure()
ax = fig.add_axes([0,0,1,1])# graph is plot betwwen a rwectangle inclosed in (0,0) and (1,1)
language = ['English','hindi','japaness','german','french']
people = [100,50,36,78,56]
ax.bar(language,people)
plt.xlabel('language')
plt.ylabel('Number of people')
plt.show()
```

# pie chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
language = ['English','hindi','japaness','german','french']
people = [100,50,36,78,56]
ax.pie(people,labels=language,autopct='%1.1f%%')# floating point value with one decimal degit
plt.show()
```

# scatter plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

# 3D Scatter plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,cmap='Blues')
plt.show()
```
