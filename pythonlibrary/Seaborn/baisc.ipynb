```python
# importing the library
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

#Note: Seaborn has some built-in datasets

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')#importing tips data from seaborn 
```

```python
tips.head()
```

```python
#setting a theme for the plot
sns.set_theme() 
```

```python
# we are going to visulaize this data 
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
# load the iris dataset 
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
# loading the titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
#house price dataset
from sklearn.datasets import  load_boston
house_boston = load_boston()# this data will be on numpy format

#convert numpy format to pandas dataframe
house = pd.DataFrame(house_boston.data,colums=house_boston.feature_name)
house['PRICE'] = house_boston.target
```

```python
print(house_boston)
```

```python
house.head()
```

# distribution plot

```python
sns.distplot(house['price'])
```

# correlation: 
positive and negative

#heatmap

```python
correlation = house.corr()
```

```python
#constructing a Heat Map
plt.figure(gigsize=(10,10))
sns.heatmap(correlation , cbar = true , square = true , fmt = '.1f',annot = true , annot_kws ={'size':8} , cmap = "blues" )
```
